```python
df = pd.read_csv("School Basic Details_Sample_Data.csv")
class_columns = [
    'Students in Class I (i_students)',
    'Students in Class II (ii_students)',
    'Students in Class III (iii_students)',
    'Students in Class IV (iv_students)',
    'Students in Class V (v_students)',
    'Students in Class VI (vi_students)',
    'Students in Class VII (vii_students)',
    'Students in Class VIII (viii_students)',
    'Students in Class IX (ix_students)',
    'Students in Class X (x_students)',
    'Students in Class XI (xi_students)',
    'Students in Class XII (xii_students)'
]
students_per_class = df[class_columns].sum()
students_per_class.index = [col.split('(')[0].strip() for col in students_per_class.index]
print(students_per_class)
```

```python
plt.figure(figsize=(12, 6))
students_per_class_sorted = students_per_class.sort_values(ascending=True)
students_per_class_sorted.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Total Number of Students")
plt.xlabel("Class")
plt.ylabel("Number of Students")
plt.grid(axis='both', linestyle ='--')
plt.tight_layout()
plt.show()
```

```python
mean = df.groupby('School Category (school_category)')['Teachers (total_teachers)'].mean()
print(mean)
```

```python
plt.figure(figsize=(12, 6))
sorted = mean.sort_values(ascending=True)
sorted.plot(kind='barh', color='skyblue', edgecolor='black')
plt.title("Teachers Graph")
plt.xlabel("Number of Teachers")
plt.ylabel("Category")
plt.grid(axis='both', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
location_counts = df['Location Type (location_type)'].value_counts()
location_counts_df = location_counts.reset_index()
location_counts_df.columns = ['Location ', 'Number of Schools']
print(location_counts_df)
```

```python
plt.figure(figsize=(6, 5))
sns.barplot(x='Location ', y='Number of Schools', data=location_counts_df, palette='Set2')
plt.title('Number of Schools by Location')
plt.xlabel('Location')
plt.ylabel('Number of Schools')
plt.tight_layout()
plt.show()
```
