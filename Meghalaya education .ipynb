```python
District_counts = df['District (district_name)'].value_counts()

plt.figure(figsize=(8, 8))
plt.bar(District_counts.index, District_counts, color='blue')
plt.title('Count Plot of District')
plt.xlabel('District')
plt.ylabel('Count')
plt.show()
```
