```
df = pd.read_csv('/content/rainfall.csv')
month_column =[
    'Jan-Feb',
    'Mar-May',
    'Jun-Sep',
    'Oct-Dec'
]
average = df[month_column].sum()
average.index = [col.split('(')[0].strip() for col in average.index]
print(average)
```

```
average.plot(kind= "barh", figsize= (8,6), edgecolor= "black", color= "skyblue" ,grid="y")
```
