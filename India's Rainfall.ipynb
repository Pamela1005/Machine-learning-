```
df = pd.read_csv('/content/rainfall.csv')
month_column =[
    'Jan-Feb',
    'Mar-May',
    'Jun-Sep',
    'Oct-Dec'
]
average = df[month_column].mean()
average.index = [col.split('(')[0].strip() for col in average.index]
print(average)
```

```
average.plot(
    kind="bar",
    figsize=(8, 6),
    edgecolor="black",
    color="skyblue",
    grid=True ,
    title ='Average Rainfall',
    rot = 0
)
```

```
each_column=[
    'JAN',
    'FEB',
    'MAR',
    'APR',
    'MAY',
    'JUN',
    'JUL',
    'AUG',
    'SEP',
    'OCT',
    'NOV',
    'DEC'
]
avg = df[each_column].mean()
avg.index = [col.split('(')[0].strip() for col in avg.index]
print(avg)
```

```
avg.plot(
    kind="bar",
    figsize=(8, 6),
    edgecolor="black",
    color="skyblue",
    grid=True ,
    title ='Average Rainfall of Each Month',
    rot = 0
)
```
